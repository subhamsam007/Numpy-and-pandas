```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# matplotlip - plots

```python
x = np.linspace(2,10,10)
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("first plot")
plt.xlabel("x=axis")
plt.ylabel("y-axis")
```

#sub - plot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(y,x,'b')
plt.subplot(2,2,4)
plt.plot(y,x,'g')
```

# Object Oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure()
axis1 = fig.add_axes([0,0,1,1])
axis2 = fig.add_axes([0.2,0.5,0.2,0.2])
axis1.plot(x,y)
axis2.plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
