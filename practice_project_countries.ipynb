# practice question start from here


```python
# which country has the highest population
Highet_population=df[df['population']== df['population'].max()]['country']
Highet_population
```

```python
#which country has the least population
least_population=df[df['population']== df['population'].min()][['country', 'population']]
least_population
```

```python
#what is the capital of the country with least population
least_population_capital=df[df['population']== df['population'].min()]['capital_city']
least_population_capital
```

```python
#give me top 5 countries with highest democratic score
democratic_score = df.sort_values(by='democracy_score', ascending=False).head(2)["country"]
democratic_score
```

```python
#how many total regions are there
total_regions = df['region'].value_counts().count()
total_regions
```

```python
#how many countries lie in Eastern Europe region
eastern_europe_countries = df[df['region'] == 'Eastern Europe']['country']
eastern_europe_countries
```

```python
#who is the political leader of the 2nd highest populated country
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
# how many countries are there whoes political leaders are unknown
df[df['political_leader'].isna()]['country']
```

```python
#how many country have Republic in their full name **country_long row you havr to take to solve
# this is one shortcut method
df[df['country_long'].str.contains('Republic')]['country']
```

```python
# another method to find
count = 0
def find_republic(txt):
    global count
    if 'republic' in txt.lower():
        count+= 1
    return txt
df['country_long'] = df['country_long'].apply(find_republic)
print(count)
```

```python
#which country in african region has highest population
africa_df = df[df['continent'] == 'Africa']
africa_df[africa_df['population'] == africa_df['population'].max()]['country']
```
