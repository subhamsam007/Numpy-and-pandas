# now make a new coloum of episode

```python
def extract_episode(val):
    match = re.search(r'\((\d+ eps)\)', val)
    if match:
        return match.group(1)
    else:
        return ""
```

```python
df['Episode']=df['Title'].apply(extract_episode)
```

**now change the episode value  of the epiode into numeric value**

```python
df['Episode']= df['Episode'].str.replace(" eps","").astype(int)
```

# Make a new coloum for time stamp


```python
#make a new column for time stamp
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]

            return data



```

```python
df['Time Stamp'] = df['Title'].apply(extraction_time)
```

```python
import re
import numpy as np

def extract_duration_column(title):
    # First, try to find the full member count string (e.g., '474,138 members')
    member_count_match = re.search(r'(\d{1,3}(?:,\d{3})*)\s*members', title)
    if member_count_match:
        full_member_string = member_count_match.group(1) # e.g., '474,138'
        # Now, split by comma and take the last part
        parts = full_member_string.split(',')
        if len(parts) > 1:
            # Return the last part, which should be the last three digits if formatted with a comma
            return parts[-1]
    return np.nan # Return NaN if not found or not in the expected format
```

```python
df['Duration'] = df['Title'].apply(extract_duration_column)
df
```
